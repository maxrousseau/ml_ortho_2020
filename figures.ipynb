```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

import main
import os
import cv2
import numpy as np


# figure image directory
fig_dir = os.path.join(os.getcwd(), 'db/figure/')
img_labels = os.listdir(fig_dir)
img_path_list = [ os.path.join(fig_dir, i) for i in img_labels]
```

```python
# get all the landmarks for each image
ldmk_list = []

for image in img_path_list:
    img_no, ldmks = main.image_processing(image)
    ldmk_list.append(ldmks)
```

```python
# make a single figure illustrating the lines of each metrics drawn on
# each sample face (4x4)

# draw each face
def drawlines(IMAGE_PATH, LANDMARKS):

    ld = LANDMARKS
    im_fi= cv2.cvtColor(cv2.imread(IMAGE_PATH), cv2.COLOR_BGR2RGB)
    im_lfh = cv2.cvtColor(cv2.imread(IMAGE_PATH), cv2.COLOR_BGR2RGB)
    im_mwfh = cv2.cvtColor(cv2.imread(IMAGE_PATH), cv2.COLOR_BGR2RGB)
    im_mwfw = cv2.cvtColor(cv2.imread(IMAGE_PATH), cv2.COLOR_BGR2RGB)
    
    color1 = (37, 167, 176)
    color2 = (169, 37, 176)
    
    im_fi = cv2.line(im_fi, (ld[27][0], ld[27][1]), (ld[8][0], ld[8][1]), color1, 15)
    im_fi = cv2.line(im_fi, (ld[1][0], ld[1][1]), (ld[15][0], ld[15][1]), color2, 15)

    
    im_lfh = cv2.line(im_lfh, (ld[33][0], ld[33][1]), (ld[8][0], ld[8][1]), color1, 15)
    im_lfh = cv2.line(im_lfh, (ld[27][0], ld[27][1]), (ld[8][0], ld[8][1]), color2, 15)
  
    im_mwfh = cv2.line(im_mwfh, (ld[4][0], ld[4][1]), (ld[12][0], ld[12][1]), color1, 15)
    im_mwfh = cv2.line(im_mwfh, (ld[27][0], ld[27][1]), (ld[8][0], ld[8][1]), color2, 15)
  
    im_mwfw = cv2.line(im_mwfw, (ld[4][0], ld[4][1]), (ld[12][0], ld[12][1]), color1, 15)
    im_mwfw = cv2.line(im_mwfw, (ld[1][0], ld[1][1]), (ld[15][0], ld[15][1]), color2, 15)
    
    return [im_fi, im_lfh, im_mwfh, im_mwfw]

img_list = []
for i in range(len(img_path_list)):
    im_dr = drawlines(img_path_list[i], ldmk_list[i])
    img_list.append(im_dr)
```

```python
# plot the images 
img_list = np.array(img_list)
img_list = img_list.flatten()

fig = plt.figure(figsize=(37,46))
columns = 4
rows = 4
for i in range(1, columns*rows+1):
    img = img_list[i-1]
    fig.add_subplot(rows, columns, i)
    plt.imshow(img)
    plt.axis('off')
plt.show() 
```
